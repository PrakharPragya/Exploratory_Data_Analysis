# Lets's start with plotting graphs

We want to analyse score of students
* lets see the distribution of math , reading and writing scores

```python
sns.distplot(df['math score'],kde=False, hist=True,hist_kws=dict(edgecolor='k',linewidth=1))
```

```python
plt.rcParams['figure.figsize']=(20,10)
sns.countplot(df['reading score'],palette='dark')
plt.title('Reading Score',fontsize=20)
plt.show()
```

```python
plt.rcParams['figure.figsize']=(20,10)
sns.countplot(df['writing score'],palette='bright')
plt.title('Writing Score',fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.subplots_adjust(left=0.125, bottom=0.1, right=0.9, top=0.9, wspace=0.5, hspace=0.2)

plt.subplot(141)
plt.title('Gender', fontsize=20)
df['gender'].value_counts().plot.pie(autopct="%1.1f%%")

plt.subplot(142)
plt.title('Ethnicity',fontsize=20)
df['race/ethnicity'].value_counts().plot.pie(autopct="%1.1f%%")

plt.subplot(143)
plt.title('Lunch',fontsize=20)
df['lunch'].value_counts().plot.pie(autopct="%1.1f%%")

plt.subplot(144)
plt.title('Parental level of Education',fontsize=20)
df['parental level of education'].value_counts().plot.pie(autopct="%1.1f%%")
```

```python
plt.figure(figsize=(20,10))
plt.subplots_adjust(left=0.125, bottom=0.1, right=0.9, top=0.9, wspace=0.5, hspace=0.2)

plt.subplot(131)
plt.title('Math Scores', fontsize=20)
sns.barplot(x="gender", y="math score",data=df)

plt.subplot(132)
plt.title('Reading Scores', fontsize=20)
sns.barplot(x="gender", y="reading score",data=df)

plt.subplot(133)
plt.title('Writing Scores',fontsize=20)
sns.barplot(x="gender", y="writing score",data=df)

plt.show()
```

We can see that male students have scored more in maths whereas female students scored more in Reading and Writing Scores

```python
plt.figure(figsize=(15,5))
plt.subplots_adjust(left=0.125, bottom=0.1, right=0.9, top=0.9, wspace=0.5, hspace=0.2)

plt.subplot(131)
plt.title('Math Scores',fontsize=20)
sns.barplot(hue="gender", y='math score', x='test preparation course',data=df)

plt.subplot(132)
plt.title('Reading Scores',fontsize=20)
sns.barplot(hue="gender", y='reading score', x='test preparation course',data=df)

plt.subplot(133)
plt.title('Writing Scores',fontsize=20)
sns.barplot(hue="gender", y='writing score', x='test preparation course',data=df)

plt.show()
```

We see the relation between test preparation course and gender for different scores

It is clear that students who have completed test preparation course have scored more in all the 3 subjects

```python
plt.figure(figsize=(15,5))
plt.subplots_adjust(left=0.125, bottom=0.1, right=0.9, top=0.9, wspace=0.5, hspace=0.2)

plt.subplot(131)
plt.title('Math Scores',fontsize=20)
sns.barplot(x='race/ethnicity', y='math score', hue='test preparation course', data=df)

plt.subplot(132)
plt.title('Reading Scores',fontsize=20)
sns.barplot(x='race/ethnicity', y='writing score', hue='test preparation course', data=df)

plt.subplot(133)
plt.title('Writing Scores',fontsize=20)
sns.barplot(x='race/ethnicity', y='writing score', hue='test preparation course', data=df)

plt.show()
```

Now we look at the relation between test preparation courses and race/ethnicity.
* overall group A has scored the least
* the difference between completed and none completed courses is also high for group A 

It is also observed that students of group E have scored higher in all three subjects

```python
plt.figure(figsize=(20,10))
plt.subplots_adjust(left=0.125, bottom=0.1, right=0.9, top=0.9, wspace=0.5, hspace=0.2)
plt.subplot(141)
plt.title('Test Preparation course Vs Gender')
sns.countplot(x='gender',hue='test preparation course', data=df)
plt.subplot(142)
plt.title('Test Preparation course Vs Ethnicity')
sns.countplot(y='race/ethnicity',hue='test preparation course', data=df)
plt.subplot(143)
plt.title('Test Preparation course Vs Lunch')
sns.countplot(x='lunch',hue='test preparation course', data=df)
plt.subplot(144)
plt.title('Test Preparation course Vs Parental Education')
sns.countplot(y='parental level of education',hue='test preparation course', data=df)
plt.show()
```

## Observations

* Most of the students have not completed the test preparation course.
* Highest number of students who belong to group C ethnicity have completed the test preparation course.
* Standard lunch students have completed the test preparation course.
* Students whose parental level of education is of "some college" has the highest difference between those who completed the course and those who did not

```python
sum_column=df['math score']+df['reading score']+df['writing score']
df['total marks']=sum_column
df['percentage']=sum_column/3
df.loc[(df['total marks']>240) & (df['total marks']<301),'grades']='A'
df.loc[(df['total marks']>210) & (df['total marks']<241),'grades']='B'
df.loc[(df['total marks']>180) & (df['total marks']<211),'grades']='C'
df.loc[(df['total marks']>150) & (df['total marks']<181),'grades']='D'
df.loc[(df['total marks']>0) & (df['total marks']<151),'grades']='E'
        
df.info()
        
```

Now total marks, percentage and grades column has been created

The grade distribution is as follows:
* A: 80-100 %
* B: 70-80 %
* C: 60-70 %
* D: 50-60 %
* E: 50 % and below
Here the upper range is inclusive whereas th lower raange is not.

Grade distribution will provide us with a better analysis of all the scores on different variables.

```python
df['grades'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

We can infer from this data that most of the students have got grades B, C which lie in the percentage range of 60-80 %.

```python
df.percentage.median()
```

```python
sns.boxplot(df['percentage'])
```

To understand a better analysis of percentages among students we use the box plot which tells the average median and low scores in a more systematic way

```python
plt.figure(figsize=(30,10))
plt.subplots_adjust(left=0.125,bottom=0.1, right=0.9, top=0.9, wspace=0.5, hspace=0.2)

plt.subplot(251)
plt.title('Grades and Gender')
sns.countplot(hue='gender', x='grades', data=df)

plt.subplot(252)
plt.title('Grades and Lunch')
sns.countplot(hue='lunch', x='grades', data=df)

plt.subplot(253)
plt.title('Grades and Test Preparation Course')
sns.countplot(hue='test preparation course', x='grades', data=df)

plt.show()
```

## Observations:

* While the higher grades such as A and B have been obtained by female students, the lower grades has a higher percentage of male students
* Free/ reduced lunch takers has lower grades specially in E and D.
* Most of the students who obtained A grade have taken the test preparation course.

```python
plt.figure(figsize=(30,10))
plt.title('Grades and Parental level of Education',fontsize=20)
sns.countplot(hue='grades', x='parental level of education', data=df)
plt.show()
```

```python
plt.figure(figsize=(40,10))
gr=pd.crosstab(df['grades'],df['race/ethnicity'],normalize=0)
gr.plot.bar(stacked=True)
plt.title('Grades and Ethnicity', fontsize=20)
plt.show()
```

## Observations:

* students whose parental level of education is of master's degree has relatively faired better and have no E grade.
* Group E has faired better than other groups.
